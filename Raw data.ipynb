```python
import torch
import mne
from matplotlib import pyplot as plt
import torch.utils.data as utils
```

```python
def splitTest(full_dataset, ratio=0.8):
    train_size = int(ratio * len(full_dataset))
    test_size = len(full_dataset) - train_size
    train_dataset, test_dataset = torch.utils.data.random_split(full_dataset, [train_size, test_size])
    return train_dataset, test_dataset

def loadDataToTensor(raw):
    data = raw[:][0]
    return torch.stack([torch.Tensor(i) for i in data])
```

```python
fileName = './test_5s_seg/CPSZ/8544_s4_t11_se2_15_raw.fif'
raw = mne.io.read_raw_fif(fileName)
loadDataToTensor(raw)
```

```python
import os
loadFiles = {}

for root, dirs, files in os.walk("./seizure"):  
    for filename in files:
        dest = root + '/' + filename
        raw = mne.io.read_raw_fif(dest)
        loadFiles[dest] = {}
        loadFiles[dest]['data'] = loadDataToTensor(raw)
        loadFiles[dest]['label'] = 1

for root, dirs, files in os.walk("./noseizure"):  
    for filename in files:
        dest = root + '/' + filename
        raw = mne.io.read_raw_fif(dest)
        loadFiles[dest] = {}
        loadFiles[dest]['data'] = loadDataToTensor(raw)
        loadFiles[dest]['label'] = 0

torch.save(loadFiles, "loadFile.data")
```
